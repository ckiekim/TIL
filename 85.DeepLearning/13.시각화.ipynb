```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

```python
def f(params):
    x, y = params['x'], params['y']
    return x*x / 20.0 + y*y

def df(params):
    x, y = params['x'], params['y']
    return {'x':x / 10.0, 'y':2*y}
```

```python
params = {'x':-7.0, 'y':2.0}
grads = df(params)
print(f'초기값: params={params}, grads={grads}, func={f(params):.4f}')
```

```python
plt.figure(figsize=(8,8))
x = -2
for i in range(1,9):
    ellipse = Ellipse(xy=(0,0), width=np.sqrt(i*4), height=np.sqrt(i*0.4), 
                      color='b', angle=0, fill=False)
    plt.gca().add_patch(ellipse)
    n = 9-i
    y = np.sqrt(0.4*n*(1-x*x/(n*4)))
    plt.plot((x,0), (y,0))
    print(f'({x},{y})')
plt.axhline(0, color='gray', lw=1), plt.axvline(0, color='gray', lw=1)
plt.axis('square')
plt.xlim(-3,3), plt.ylim(-3,3)
plt.show()
```
