### 2. 베르누이분포와 이항분포

```python
import numpy as np
import scipy as sp
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['font.family'] = 'Malgun Gothic'
```

- 베르누이 분포

```python
mu = 0.6
rv = sp.stats.bernoulli(mu)
```

```python
xx = [0, 1]
plt.bar(xx, rv.pmf(xx))         # 확률질량함수: probability mass function
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["x=0", "x=1"])
plt.xlabel("표본값")
plt.ylabel("P(x)")
plt.title("베르누이 확률변수의 확률질량함수")
plt.show()
```

```python
x = rv.rvs(100, random_state=2022)
x
```

```python
sns.countplot(x)
plt.title("베르누이 확률변수의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.show()
```

```python
np.mean(x), np.var(x)
```

```python
sp.stats.describe(x)
```

- 이항 분포

```python
N = 10
mu = 0.6
rv = sp.stats.binom(N, mu)
```

```python
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")
plt.show()
```

```python
x = rv.rvs(100, random_state=2022)
x
```

```python
sns.countplot(x)
plt.title("이항분포의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.show()
```

```python
# 연습 문제 8.2.3
# 1) μ=0.5, N=5
N, mu = 5, 0.5
rv = sp.stats.binom(N, mu)
x = rv.rvs(10, random_state=2022)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")

plt.subplot(122)
sns.countplot(x)
plt.title("이항분포의 시뮬레이션 결과")
plt.xlabel("표본값")

plt.show()
```

```python
# 2) μ=0.9, N=20
N, mu = 20, 0.9
rv = sp.stats.binom(N, mu)
x = rv.rvs(1000, random_state=2022)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")

plt.subplot(122)
sns.countplot(x)
plt.title("이항분포의 시뮬레이션 결과")
plt.xlabel("표본값")

plt.show()
```

### 3. 카테고리분포와 다항분포

- 카테고리 분포

```python
mu = [0.1, 0.1, 0.1, 0.1, 0.3, 0.3]
rv = sp.stats.multinomial(1, mu)
```

```python
xx = np.arange(1, 7)
xx_ohe = pd.get_dummies(xx)

plt.bar(xx, rv.pmf(xx_ohe.values))
plt.ylabel("P(x)")
plt.xlabel("표본값")
plt.title("카테고리분포의 확률질량함수")
plt.show()
```

```python
X = rv.rvs(100, random_state=100)
X[:5]
```

```python
y = X.sum(axis=0) / float(len(X))
plt.bar(np.arange(1, 7), y)
plt.title("카테고리분포의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.ylabel("비율")
plt.show()
```

- 다중 분류

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["품종"] = pd.Series(iris.target, dtype="category")

df1 = df[df["petal width (cm)"] > 1.5]
df2 = df[df["petal width (cm)"] <= 1.5]

fig, ax = plt.subplots(1, 2)
sns.countplot(x="품종", data=df1, ax=ax[0]).set_title("꽃잎 폭 > 1.5cm")
sns.countplot(x="품종", data=df2, ax=ax[1]).set_title("꽃잎 폭 <= 1.5cm")
plt.tight_layout()
plt.show()
```

- 다항분포(Multinomial distribution)

```python
N = 30
mu = [0.1, 0.1, 0.1, 0.1, 0.3, 0.3]
rv = sp.stats.multinomial(N, mu)

X = rv.rvs(100, random_state=2022)
X[:5]
```

```python
df = pd.DataFrame(X).stack().reset_index()
df.columns = ["시도", "클래스", "표본값"]
df.head()
```

```python
sns.violinplot(x="클래스", y="표본값", data=df, inner="quartile")
sns.swarmplot(x="클래스", y="표본값", data=df, color=".3")
plt.title("다항분포의 시뮬레이션 결과")
plt.show()
```

```python
plt.boxplot(X)
plt.title("다항분포의 시뮬레이션 결과")
plt.xlabel("클래스")
plt.ylabel("표본값")
plt.show()
```
