### Image Filtering
- 고주파를 제거하면 Blur 처리
- 저주파를 제거하면 대상의 영역 확인 

```python
# https://pixabay.com/ko/photos/%eb%b2%8c-%ea%b3%a4%ec%b6%a9-%ea%bd%83-%ea%bf%80%eb%b2%8c-%ec%88%98%eb%b6%84-5618012/
uploaded = files.upload()
```

```python
img = cv2.imread('bee.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img.shape
```

```python
plt.imshow(img);
```

```python
# Low Pass Filter
kernel = np.ones((5,5), np.float32) / 25
kernel
```

```python
dst = cv2.filter2D(img, -1, kernel)
plt.figure(figsize=(12,5))
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(dst), plt.title('After Low-Pass filter')
plt.show()
```

- Averaging

```python
dst1 = cv2.blur(img, ksize=(7,7))
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(dst1), plt.title('Averaging Blurring')
plt.show()
```

- Gaussian filtering

```python
dst2 = cv2.GaussianBlur(img, ksize=(5,5), sigmaX=0)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(dst2), plt.title('Gaussian filter blurring')
plt.show()
```

- Median filtering: salt-and-pepper noise 제거

```python
dst3 = cv2.medianBlur(img, ksize=9)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(dst3), plt.title('Median blurring')
plt.show()
```

- Bilateral filtering

```python
dst4 = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(dst4), plt.title('Bilateral blurring')
plt.show()
```

- 종합 비교

```python
images = [img, dst, dst1, dst2, dst3, dst4]
titles = ['Original', 'Low-Pass Filter(5x5)', 'Blur(7x7)',
          'Gaussian Blur(5x5)', 'Median Blur(9)', 'Bilateral(9)']

plt.figure(figsize=(18,9))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
