```python
img = cv2.imread('travel.jpg', 0)
img.shape
```

```python
plt.imshow(img, cmap='gray');
```

- Sobel

```python
sobelx = cv2.Sobel(img, ddepth=cv2.CV_8U, dx=1, dy=0, ksize=3)
plt.imshow(sobelx, cmap='gray');
```

```python
sobely = cv2.Sobel(img, ddepth=cv2.CV_8U, dx=0, dy=1, ksize=3)
plt.imshow(sobely, cmap='gray');
```

```python
sobelx2 = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
plt.imshow(sobelx2, cmap='gray');
```

```python
sobely2 = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
plt.imshow(sobely2, cmap='gray');
```

- Laplacian

```python
laplacian = cv2.Laplacian(img, cv2.CV_8U)
plt.imshow(laplacian, cmap='gray');
```

```python
laplacian2 = cv2.Laplacian(img, cv2.CV_64F)
plt.imshow(laplacian2, cmap='gray');
```

```python
canny = cv2.Canny(img, 30, 70)
plt.imshow(canny, cmap='gray');
```
