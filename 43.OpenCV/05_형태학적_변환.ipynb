```python
origin = cv2.imread('morph_origin.png')
dot = cv2.imread('morph_dot.png')
hole = cv2.imread('morph_hole.png')
```

```python
plt.figure(figsize=(10,4))
plt.subplot(131), plt.imshow(origin), plt.title('Original')
plt.subplot(132), plt.imshow(dot), plt.title('Dot')
plt.subplot(133), plt.imshow(hole), plt.title('Hole')
plt.show()
```

- Rectangle kernel

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
erosion = cv2.erode(dot, kernel, iterations=1)
dilation = cv2.dilate(hole, kernel, iterations=1)

opening = cv2.morphologyEx(dot, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(hole, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(origin, cv2.MORPH_GRADIENT, kernel)
tophat = cv2.morphologyEx(origin, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(origin, cv2.MORPH_BLACKHAT, kernel)
```

```python
images = [dot, erosion, opening, hole, dilation, closing, gradient, tophat, blackhat]
titles = ['dot', 'erosion', 'opening', 'hole', 'dilation', 'closing', 
          'gradient(dilation - erosion)', 'tophat(origin vs opening)', 'blackhat(origin vs closing)']
```

```python
plt.figure(figsize=(10,12))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
